```
# =========================================
# GENERALIDADES DE PYTHON
# Código básico para usar en Google Colab
# =========================================


# -----------------------------------------
# 1. IMPRIMIR TEXO EN PANTALLA
# -----------------------------------------

print("Hola mundo")  # Muestra texto en consola


# -----------------------------------------
# 2. VARIABLES
# -----------------------------------------

nombre = "Elizabeth"   # Variable tipo texto
edad = 30              # Variable tipo entero
altura = 1.73          # Variable decimal
es_mayor = True        # Variable booleana

print(nombre)
print(edad)
print(altura)
print(es_mayor)


# -----------------------------------------
# 3. TIPOS DE DATOS
# -----------------------------------------

texto = "Python"       # String
numero = 10            # Integer
decimal = 3.14         # Float
booleano = False       # Boolean

print(type(texto))     # Muestra el tipo de dato


# -----------------------------------------
# 4. OPERACIONES MATEMÁTICAS
# -----------------------------------------

a = 10
b = 5

print(a + b)   # Suma
print(a - b)   # Resta
print(a * b)   # Multiplicación
print(a / b)   # División
print(a ** b)  # Potencia
print(a % b)   # Módulo o residuo


# -----------------------------------------
# 5. ENTRADA DE DATOS
# -----------------------------------------

nombre_usuario = input("Ingresa tu nombre: ")

print("Hola", nombre_usuario)


# -----------------------------------------
# 6. CONDICIONALES
# -----------------------------------------

edad = 18

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")


# -----------------------------------------
# 7. CONDICIONALES MÚLTIPLES
# -----------------------------------------

nota = 4.5

if nota >= 4.5:
    print("Excelente")
elif nota >= 3:
    print("Aprobaste")
else:
    print("Reprobaste")


# -----------------------------------------
# 8. CICLO FOR
# -----------------------------------------

for i in range(5):   # Repite de 0 a 4
    print(i)


# -----------------------------------------
# 9. CICLO WHILE
# -----------------------------------------

contador = 0

while contador < 5:
    print(contador)
    contador += 1


# -----------------------------------------
# 10. LISTAS
# -----------------------------------------

frutas = ["Manzana", "Pera", "Banano"]

print(frutas)
print(frutas[0])   # Primer elemento

frutas.append("Uva")   # Agrega elemento

print(frutas)


# -----------------------------------------
# 11. TUPLAS
# -----------------------------------------

colores = ("Rojo", "Azul", "Verde")

print(colores)


# -----------------------------------------
# 12. DICCIONARIOS
# -----------------------------------------

persona = {
    "nombre": "Elizabeth",
    "edad": 30,
    "ciudad": "Medellín"
}

print(persona["nombre"])


# -----------------------------------------
# 13. FUNCIONES
# -----------------------------------------

def saludar(nombre):
    return "Hola " + nombre

mensaje = saludar("Elizabeth")

print(mensaje)


# -----------------------------------------
# 14. FUNCIONES CON PARÁMETROS
# -----------------------------------------

def sumar(a, b):
    return a + b

resultado = sumar(5, 10)

print(resultado)


# -----------------------------------------
# 15. MANEJO DE ERRORES
# -----------------------------------------

try:
    numero = int(input("Ingresa un número: "))
    print(numero)
except:
    print("Error: Debes ingresar un número")


# -----------------------------------------
# 16. IMPORTAR LIBRERÍAS
# -----------------------------------------

import math   # Librería matemática

print(math.sqrt(25))   # Raíz cuadrada


# -----------------------------------------
# 17. CLASES Y OBJETOS
# -----------------------------------------

class Persona:

    # Constructor
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    # Método
    def mostrar(self):
        print(self.nombre, self.edad)


# Crear objeto
persona1 = Persona("Elizabeth", 30)

persona1.mostrar()


# -----------------------------------------
# 18. LIST COMPREHENSION
# -----------------------------------------

numeros = [1, 2, 3, 4, 5]

cuadrados = [x**2 for x in numeros]

print(cuadrados)


# -----------------------------------------
# 19. ARCHIVOS
# -----------------------------------------

# Crear archivo
archivo = open("texto.txt", "w")

archivo.write("Hola Python")

archivo.close()


# Leer archivo
archivo = open("texto.txt", "r")

print(archivo.read())

archivo.close()


# -----------------------------------------
# 20. COMENTARIOS
# -----------------------------------------

# Comentario de una línea

"""
Comentario
de varias líneas
"""


# -----------------------------------------
# 21. F-STRINGS
# -----------------------------------------

nombre = "Elizabeth"
edad = 30

print(f"Mi nombre es {nombre} y tengo {edad} años")


# -----------------------------------------
# 22. OPERADORES LÓGICOS
# -----------------------------------------

x = True
y = False

print(x and y)   # AND
print(x or y)    # OR
print(not x)     # NOT


# -----------------------------------------
# 23. CONVERTIR DATOS
# -----------------------------------------

numero_texto = "10"

numero_entero = int(numero_texto)

print(numero_entero)


# -----------------------------------------
# 24. RECORRER LISTAS
# -----------------------------------------

nombres = ["Ana", "Luis", "Carlos"]

for nombre in nombres:
    print(nombre)


# -----------------------------------------
# 25. ENUMERATE
# -----------------------------------------

for indice, valor in enumerate(nombres):
    print(indice, valor)


# -----------------------------------------
# FIN DEL ARCHIVO
# -----------------------------------------
```
