```python
# Pandas

# Importar pandas

import pandas as pd
```

```python
# 1. importar JSON

url = "https://www.datos.gov.co/resource/3t73-n4q9.json"

df_json = pd.read_json(url)
print(df_json)
```

```python
# Movimiento N1

# Ver cuántas filas tiene

df_json.shape
```

```python
# Movimiento N2

# Ver las primeras 5 filas

print(df_json.head())
```

```python
# Movimiento N3

# Ver las ultimas 5 filas

print(df_json.tail())
```

```python
# Movimiento N4

# Informacion del dataframe

print(df_json.info())
```

```python
# Movimiento N5

# Ver nombres de columnas

print(df_json.columns)
```

```python
# Movimiento N6

# Estadistica basica

print(df_json.describe())
```

```python
# Movimiento N7

# Ver indices

print(df_json.index)
```

```python
# Movimiento N8

# Seleccionar una columna

print(df_json["principio_activo"])
```

```python
# Movimiento N9

# Seleccionar varias columnas

print(df_json[["principio_activo", "concentracion", "nombre_comercial"]])
```

```python
# Movimiento N10

# Selecciionar una fila

print(df_json.iloc[0])
```

```python
# Movimiento N11

# Seleccionar filas con columnas

print(df_json.iloc[3:8, 1:4])
```

```python
# Movimiento N12

# Filtrar datos

factor_3 = df_json[df_json["numerofactor"] == 3]
print(factor_3)
```

```python
# Movimiento N13

# Eliminar columnas

df_json = df_json.drop("fabricante", axis=1)
print(df_json)
```

```python
# Movimiento N14

# Eliminar filas

df_json = df_json.drop(1, axis=0)
print(df_json)
```

```python
# Movimiento N15

# Ordenar datos

order = df_json.sort_values("precio_por_tableta")
print(order)
```

```python
# Movimiento N16

# Ordenar descendiente

order_desc = df_json.sort_values("precio_por_tableta", ascending=False)
print(order_desc)
```

```python
# Movimiento N17

# Cambiar el nombre de las columnas

df_json = df_json.rename(columns={"principio_activo": "active_ingredient", "concentracion": "concentration", "nombre_comercial": "brand_name", "precio_por_tableta": "price_per_tablet", "numerofactor": "factor_number"})
print(df_json)
```

```python
# Movimiento N18

# Contar registros

count = df_json["active_ingredient"].value_counts()
print(count)
```

```python
# Movimiento N19

# Recorrer filas

for index, row in df_json.iterrows():
    print(index, row["active_ingredient"])
```

```python
# Movimiento N20

# Guardar en csv

df_json.to_csv("medicamentos.csv", index=False)
```

```python
# Movimiento N21

# Leer un archivo csv

new_df = pd.read_csv("medicamentos.csv")
print(new_df)
```
